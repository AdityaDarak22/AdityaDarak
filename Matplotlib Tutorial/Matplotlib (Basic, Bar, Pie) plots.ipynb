## Horizontal Bar Charts


```python
# Creating a bar chart of most popular programming language 
# data used for this will be read from a csv file
```

```python
# File be read using csv module as well using pandas csv method
import csv
from collections import Counter
```

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_count =Counter() #Counts the number of occurences of the language appeared
    for row in csv_reader:
        language_count.update(row['LanguagesWorkedWith'].split(';'))
        
print(language_count.most_common(15)) # list the n most elements from the count
        
```

```python
# Now we want the languages in one list and the popularity numbers in the other list
languages=[]
popularity=[]
for item in language_count.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
print(languages)
print('\r')
print(popularity)
```

```python
# plotting the bar chart
plt.bar(languages,popularity)
plt.xlabel("Programming languages")
plt.ylabel("No of people using the languages")
plt.title("Most Popular Language")
plt.show()

# its difficult to read all the x values in the bar chart
# printing a horizontal bar chart is the better option in case we have more values in xaxis
```

```python
languages.reverse()
popularity.reverse()

plt.barh(languages,popularity)
plt.xlabel("Programming languages")
plt.ylabel("No of people using the languages")
plt.title("Most Popular Language")
plt.show()
```

```python
# Loading the same data using pandas which is much more faster
```

```python
data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang = data['LanguagesWorkedWith']
lang_counter = Counter()
for response in lang:
    lang_counter.update(response.split(';'))
lang1 = []
popu1 = []
for item in lang_counter.most_common(15):
    lang1.append(item[0])
    popu1.append(item[1])
    
lang1.reverse()
popu1.reverse()

plt.barh(lang1,popu1,alpha=0.5)
plt.xlabel("Programming languages")
plt.ylabel("No of people using the languages")
plt.title("Most Popular Language")
plt.show()
```
