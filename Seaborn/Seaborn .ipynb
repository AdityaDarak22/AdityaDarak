```python
# importing the library
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore',category =FutureWarning)
```

```python
# when we our comparing one features like f1 then this is called as univariate analysis
# when we our comparing two features like f1, f2 then this is called as bivariate analysis

# this can be analyzed using seaborn visualisation library
```

```python
# distplot

# jointplot : is used the most bivariate analysis

# pairplot is considered when more than two features are present
```

```python
# Loading a dataframe
df = sns.load_dataset("tips") # Restaurent dataset
```

```python
df.head()

# tip is the dependent feature
# rest all columns are independent feature
```

```python
df.info()
```

### Correlation with Heatmap

- A correlation map uses colored cells to show a correaltion matrix between two discrete dimensions or event types. A very important step in feature selection.
- correlation is always find between the integer variables.

```python
df.corr()

# if total bill increases then tip will also increase.
# if the correlation was negative then as the bill is very low then tip given is also very less
```

```python
sns.heatmap(df.corr())
plt.show()
```

### Jointplot

- it allows to understand the relation between two variables. the central chart display the correlation.
- it is usually as scatter plot, a hexbin plot, a 2D Histogram, 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
plt.show()

# by using reg it will draw a kde or a pdf graph
```

### Pairplot

   - it is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value
   - all the variables are paired with other variables

```python
sns.pairplot(df)
plt.show()
```

```python
# if we want to classify the features based on sex 
# then we pass sex in the hue parameter of seaborn

sns.pairplot(df, hue='sex')
plt.show()
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df, hue='smoker')
plt.show()
```

### Distplot
- it helps us to check the distribution of the columns feature.


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False,bins=10)
plt.show()
```

## Categorical Plots

- boxplot

```python
sns.barplot(y='total_bill',x='smoker',data=df)
plt.show()
```

### Box Plot
 - a box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five number summary

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
plt.show()
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=df, palette='rainbow')
plt.show()
```

```python
sns.boxplot(data=df,orient='v')
plt.show()
```

```python
sns.boxplot(y='day',x= 'total_bill',hue='smoker',data=df)
plt.show()
```

### Violin plot
- this plot helps us to see the distribution of data in terms of kernal density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
plt.show()
```
